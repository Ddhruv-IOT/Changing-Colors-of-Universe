```python
def show():
    global m 
    
    label.config( text = clicked.get() )
    m = eval("cv2." + f"{clicked.get()}")

cam = cv2.VideoCapture(0)
m = cv2.COLOR_RGB2GRAY
cv2_based_image = " "

def video_camera(): 
    global cv2_based_image
    
    r, frame = cam.read()
    cv2_based_image = cv2.cvtColor(frame, m)
    
    image = Image.fromarray(cv2_based_image)
    photo = ImageTk.PhotoImage(image=image)
    
    label.imgtk = photo
    
    label.configure(image=photo)
    label.after(10, video_camera) # keep looping to get video
    
def save():
    name = random.random()
    global cv2_based_image
    cv2.imwrite(f"{name}.png", cv2_based_image) 
```

```python
root = tk.Tk() # master window 

root.geometry("900x700")
root.minsize(100, 100)
root.maxsize(800, 700)

bg = tk.PhotoImage(file="dg.png")

label = tk.Label(root, image=bg) 
label.place(x=0, y=0)

clicked = tk.StringVar()
clicked.set( "COLOR_RGB2GRAY" )

tk.Label(root, text="Be Ready....", font=("times new roman", 15, "bold")).pack()

drop = tk.OptionMenu( root , clicked , *modes )
drop.pack(pady=3)

tk.Button( root , text = "click Me" , command = show ).pack(pady=3)
tk.Button( root , text = "Save Pic" , command = save).pack(pady=3)

img_frame = tk.Frame(root, width=200, height=200) # frame to hold label
img_frame.pack(pady=20)
label = tk.Label(img_frame, text="Be Ready....", font=("times new roman", 15, "bold")) # label inside frame to hold image output of CV2
label.pack()
```

```python
video_camera() # start camera function
root.mainloop() # start GUI
cam.release() # on close camera will be released and turned off.
```
